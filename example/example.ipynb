```python
import matplotlib.pyplot as plt
import numpy as np
import yaml
```

```python
with open('./mastermao-simtimes.mplstyle', 'r', encoding='utf-8') as stream:
    style = yaml.load(stream, Loader=yaml.Loader)
plt.style.use(style)
```

```python
def model(x, p):
    return x ** (2 * p + 1) / (1 + x ** (2 * p))
```

```python
x = np.linspace(0.75, 1.25, 201)

fig, ax = plt.subplots(figsize=(7.05, 3), constrained_layout=True)
for p in [5, 7, 10, 15, 20, 30, 38, 50, 100]:
    ax.plot(x, model(x, p), label=p)
ax.legend(title='Order', fontsize=7)
ax.set(xlabel=r'电压 (mV)')
ax.set(ylabel=r'电流 ($\mu$A)')
ax.autoscale(tight=True)
plt.show()
# fig.savefig('../images/fig.webp')
```
